```python
import urllib
import json
import pandas as pd

url = 'http://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/1.0_week.geojson'
d = json.loads(urllib.urlopen(url).read())

data = pd.DataFrame(d.items())

data
```
